```python
from sympy import symbols, diff

x, y = symbols('x y')

expr = x**2 + y**2 + 3*y + 2*x

# derivative w.r.t x
dx = diff(expr, x)

# derivative w.r.t y
dy = diff(expr, y)

print("d/dx:", dx)
print("d/dy:", dy)
```
